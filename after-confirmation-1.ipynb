### ReliefF

```python
import sklearn_relief as sr
import numpy as np

r = sr.RReliefF(n_features = 2, n_jobs=1)
r.fit(feature_data_no_row.iloc[:, 1:10].to_numpy(), label_data.to_numpy())

# rrelieff takes long time to run

print(r.w_)

feat_indices = np.flip(np.argsort(r.w_), 0)[0:5]
print(feat_indices)

```

```python
feature_data_no_row.iloc[:, 1:10].iloc[:, feat_indices]
```

```python
def relieff_select(X: pd.DataFrame, y: pd.Series, k: int, n_jobs=1):
    if n_jobs == -1: 
        r = sr.RReliefF(n_features = k)
    else:
        r = sr.RReliefF(n_features = k, n_jobs=n_jobs)
    r.fit(X.to_numpy(), y.to_numpy())
    feat_indices = np.flip(np.argsort(r.w_), 0)[0:k]
    return feat_indices, r.w_[feat_indices]
```

```python
indices, scores = relieff_select(feature_data_no_row, label_data, 5)

print(indices)
print(scores)
```

### Embedded Selection (Using Elastic Net)

```python
# import elastic net 

from sklearn.linear_model import ElasticNet

# import matplotlib

import matplotlib.pyplot as plt

# instantiate model

enet = ElasticNet(max_iter=10000)

# fit model

enet.fit(feature_data_no_row, label_data)

# take out the coefficients

enet_coef = enet.coef_

# plot the coefficients as a bar plot

# plt.figure(figsize=(15, 7))

# plt.plot(range(feature_data_no_row.shape[1]), enet_coef)

enet_coef_abs = np.abs(enet_coef)

feature_indices = np.flip(np.argsort(enet_coef_abs), 0)[0:5]

feature_indices
```

```python
def enet_select(X: pd.DataFrame, y: pd.Series, k: int, max_iter=10000):
    enet = ElasticNet(max_iter=max_iter)
    enet.fit(X,y)
    coef = enet.coef_
    abs_coef = np.abs(coef)
    indices = np.flip(np.argsort(abs_coef), 0)[0:k]
    return indices, coef[indices]
    
```

```python
indices, scores = enet_select(feature_data_no_row, label_data, 5)

# Print selected features

print(indices)
print(scores)
```

### Embedded Selection (Random Forest)

```python
# import random forest 

from sklearn.ensemble import RandomForestRegressor

# import matplotlib
import numpy as np 
import matplotlib.pyplot as plt

# instantiate model

rf = RandomForestRegressor(n_estimators=1000)

# train the model on training data

rf.fit(feature_data_no_row, label_data)

# extract feature importances

rf_coef = rf.feature_importances_

# plot the coefficients as a bar plot

plt.figure(figsize=(15, 7))

plt.plot(range(feature_data_no_row.shape[1]), rf_coef)

rf_coef_abs = np.abs(rf_coef)

feature_indices = np.flip(np.argsort(rf_coef_abs), 0)[0:5]

feature_indices
```

```python
from sklearn.ensemble import RandomForestRegressor


def rf_select(X: pd.DataFrame, y: pd.Series, k: int, **kwargs):
    rf = RandomForestRegressor(**kwargs)
    rf.fit(X, y)
    coef = rf.feature_importances_
    indices = np.argsort(coef)[::-1][:k]
    return indices, coef[indices]
```
