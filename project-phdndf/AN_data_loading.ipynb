## Specialised Processing of FGFR4 Raw Simulation Data

```python
folder_path = path_loader.get_data_path()+"data\\results\\fgfr4-model-simulation-data"

loading_code = 'fgfr4_ccle_dynamic_features'
dynamic_features = data_link.get_data_from_code(loading_code) # need the row label 
```

```python
# get the row label
row_label = dynamic_features['Row']
```

```python
import pandas as pd
row_label_list = row_label.tolist()

# simulation_data = {}
simulation_data_df = pd.DataFrame()
not_found_count = 0
for cell_id in row_label_list: 
    file_name = cell_id + '.csv'
    file_path = folder_path + '\\' + file_name
    # print(file_path)
    try: 
        simulation_data = pd.read_csv(file_path)
        # rename 'Row' to 'Time'
        simulation_data.rename(columns={'Row': 'Time'}, inplace=True)
        # add the cell_id to the dataframe as a column, to the left of the 'Time' column
        simulation_data.insert(0, 'Cellline', cell_id)
        
        simulation_data_df = pd.concat([simulation_data_df, simulation_data], ignore_index=True)
        # print(f'Loaded {file_name} into simulation_data dictionary')
    except FileNotFoundError:
        # print(f'File {file_name} not found')
        not_found_count += 1

print(f'Number of files loaded: {len(simulation_data)}')
print(f'Number of files not found: {not_found_count}')

    
```

```python
import pandas as pd 

def dataset_to_sanger_model_id_from_ccle(df: pd.DataFrame, ccle_info_df: pd.DataFrame):
    '''
    Transforms DepMap_ID based-indices of the dataframe to Sanger_Model_ID
    input: df: pd.DataFrame, must have 'ACH-000000' formatted index as cell line identifiers 
    output: pd.DataFrame, with Sanger_Model_ID as index
    '''
    ccle_IDs = ccle_info_df[['DepMap_ID', 'Sanger_Model_ID']]
    ccle_IDs.set_index('DepMap_ID', inplace=True)

    transformed_df = df.join(ccle_IDs, how='inner')
    # remove rows with NaN for Sanger_Model_ID
    transformed_df.dropna(subset=['Sanger_Model_ID'], inplace=True)
    transformed_df.set_index('Sanger_Model_ID', inplace=True)
    return transformed_df
```

```python
sanger_df = dataset_to_sanger_model_id_from_ccle(fgfr4_dyn_features, ccle_sample_info)
```
