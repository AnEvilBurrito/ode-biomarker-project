## Loading CCLE Proteomics Expression Data

```python
ccle_proteomics = data_link.get_data_from_code('ccle_protein_expression')
```

```python
ccle_proteomics
```

```python
# separate the first 30 columns 

ccle_proteomics_info = ccle_proteomics.iloc[:, :48]
ccle_proteomics_processed = ccle_proteomics.iloc[:, 48:]


```

```python
ccle_proteomics_info
```

```python
# remove columns with the 'Column' string

ccle_proteomics_processed = ccle_proteomics_processed.loc[:,~ccle_proteomics_processed.columns.str.contains('Column')]

ccle_proteomics_processed
```

```python
# add the 'Gene_Symbol' column to the processed data

ccle_proteomics_processed.insert(0, 'Cell_Line', ccle_proteomics_info['Gene_Symbol'])

ccle_proteomics_processed
```

```python
# set the 'Gene_Symbol' column as the index

ccle_proteomics_processed.set_index('Cell_Line', inplace=True)
```

```python
ccle_proteomics_processed
```

```python
# transpose the data

ccle_proteomics_processed = ccle_proteomics_processed.T
```

```python
ccle_proteomics_processed
```

```python
# extract the cell line names

cell_lines = list(ccle_proteomics_processed.index)
CCLE_NAMES = []
CCLE_NAMES_dict = {}
for cell in cell_lines:
    CCLE_NAME = "_".join(cell.split("_")[:-1])
    CCLE_NAMES.append(CCLE_NAME)   
    CCLE_NAMES_dict[cell] = CCLE_NAME
```

```python
ccle_sample_info = data_link.get_data_from_code('ccle_sample_info')
```

```python
depmap_ids = {}
count = 0
found_count = 0 
for ccle_name in CCLE_NAMES:
    # get the row in the sample info data column 'CCLE_Name', and locate 'Sanger_Model_ID'
    row = ccle_sample_info.loc[ccle_sample_info['CCLE_Name'] == ccle_name]
    col_data = row['DepMap_ID'].values
    if len(col_data) > 0:
        if isinstance(col_data[0], str):
            depmap_ids[ccle_name] = col_data[0]
            found_count += 1
    count += 1
    
print(f'Found {found_count} out of {count} DepMap IDs')
```

```python
# use CCLE_NAMES_dict to map the cell line names to the Sanger Model IDs

index_row_to_depmap = {}
for key, value in CCLE_NAMES_dict.items():
    if value in depmap_ids:
        index_row_to_depmap[key] = depmap_ids[value]
```

```python
print(len(index_row_to_depmap))
```

```python
# for each cell line row in the proteomics data, replace the cell line name with the Sanger Model ID, if it exists

ccle_proteomics_processed_renamed = ccle_proteomics_processed.rename(index=index_row_to_depmap)
ccle_proteomics_processed_renamed
```

```python
# save the processed data as a pickle file, as well as ccle_proteomics_info to another pickle file

save_path = path_loader.get_data_path()+"data\\proteomic-expression\\ccle-2019-cell\\"
ccle_proteomics_processed_renamed.to_pickle(save_path+"ccle_proteomics_processed.pkl")
ccle_proteomics_info.to_pickle(save_path+"ccle_proteomics_info.pkl")
```

## Specialised Processing of FGFR4 Raw Simulation Data

```python
folder_path = path_loader.get_data_path()+"data\\results\\fgfr4-model-simulation-data"

loading_code = 'fgfr4_ccle_dynamic_features'
dynamic_features = data_link.get_data_from_code(loading_code) # need the row label 
```

```python
# get the row label
row_label = dynamic_features['Row']
```

```python
import pandas as pd
row_label_list = row_label.tolist()

# simulation_data = {}
simulation_data_df = pd.DataFrame()
not_found_count = 0
for cell_id in row_label_list: 
    file_name = cell_id + '.csv'
    file_path = folder_path + '\\' + file_name
    # print(file_path)
    try: 
        simulation_data = pd.read_csv(file_path)
        # rename 'Row' to 'Time'
        simulation_data.rename(columns={'Row': 'Time'}, inplace=True)
        # add the cell_id to the dataframe as a column, to the left of the 'Time' column
        simulation_data.insert(0, 'Cellline', cell_id)
        
        simulation_data_df = pd.concat([simulation_data_df, simulation_data], ignore_index=True)
        # print(f'Loaded {file_name} into simulation_data dictionary')
    except FileNotFoundError:
        # print(f'File {file_name} not found')
        not_found_count += 1

print(f'Number of files loaded: {len(simulation_data)}')
print(f'Number of files not found: {not_found_count}')

    
```

```python
import pandas as pd 

def dataset_to_sanger_model_id_from_ccle(df: pd.DataFrame, ccle_info_df: pd.DataFrame):
    '''
    Transforms DepMap_ID based-indices of the dataframe to Sanger_Model_ID
    input: df: pd.DataFrame, must have 'ACH-000000' formatted index as cell line identifiers 
    output: pd.DataFrame, with Sanger_Model_ID as index
    '''
    ccle_IDs = ccle_info_df[['DepMap_ID', 'Sanger_Model_ID']]
    ccle_IDs.set_index('DepMap_ID', inplace=True)

    transformed_df = df.join(ccle_IDs, how='inner')
    # remove rows with NaN for Sanger_Model_ID
    transformed_df.dropna(subset=['Sanger_Model_ID'], inplace=True)
    transformed_df.set_index('Sanger_Model_ID', inplace=True)
    return transformed_df
```
